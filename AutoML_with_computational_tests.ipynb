```python
type_of_dataset = None
r = df.shape
last_ind = df.columns.values[len(df.columns.values)-1]
total_entries = r[0]-1
set_contents = set(df[last_ind])
gandharva_constant = len(set_contents)/total_entries
if(gandharva_constant>=0.2):
  type_of_dataset = 'regression'
else:
  type_of_dataset = 'classification'
type_of_dataset
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV,RandomizedSearchCV

X = (df.drop(columns = last_ind,axis = 1))
Y = np.asarray(df[last_ind])
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2,random_state = 1)
X_train
```

```python
def self_paired_ttest(ttable_val_alpha,arr1,arr2,other):
  print("")
  print("")
  e = []
  add = 0
  add1 = 0
  for i in range(len(arr1)):

    e.append(abs(arr1[i]-arr2[i]))
    add+=(abs(arr1[i]-arr2[i]))
  sample_mean = add/len(arr1)
  for i in range(len(e)):
    add1+=((e[i]-sample_mean)**2)
  sd = (add1/(len(arr1)-1))**0.5
  t_stat = (sample_mean*(len(arr1)**0.5))/sd
  print(t_stat)
  if(-ttable_val_alpha<t_stat and t_stat<ttable_val_alpha):
    print("No significant difference between",algo_names[maxi_ind],"and",other)
    print("Since the value of paired t-test is: ",t_stat,"which falls in range(",-ttable_val_alpha,",",ttable_val_alpha,")")

  else:
    print(algo_names[maxi_ind],"performs better then ",other)
    print("Since the value of paired t-test is: ",t_stat,"which does not fall in range(",-ttable_val_alpha,",",ttable_val_alpha,")")
  print("-------------------------------------------------------------------------------------------------------------------------------")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn import tree


from sklearn.metrics import accuracy_score,confusion_matrix
if(type_of_dataset == 'classification'):
  algo_names = ['svm','random_forest','logistic_regression','decision_tree']
  keys = ['svm','random_forest','logistic_regression','decision_tree']
  algo = [svm.SVC(),RandomForestClassifier(),LogisticRegression(),tree.DecisionTreeClassifier()]
  model_params = {
   'svm':{
   'model':svm.SVC(gamma='auto'),
   'params':{
       'C':[1,10,20],
       'kernel':['linear','poly','rbf']

   }
  } ,
  'random_forest':{
       'model':RandomForestClassifier(),
       'params':{
           'n_estimators':[1,5,10,50,100,150,500,1000]
       }
   },
   'logistic_regression':{
       'model':LogisticRegression(solver = 'liblinear',multi_class='auto'),
       'params':{
          'C':[1,5,10,50,100,150,500,1000]
       }
   },
   'decision_tree':{
       'model':tree.DecisionTreeClassifier(),
       'params':{
           'criterion':['gini','entropy','log_loss'],
           'splitter':['best','random']
       }
   },
    }
  scores = []
  time = []
  for model_name,mp in model_params.items():
    clf = GridSearchCV(mp['model'],mp['params'],cv =5,return_train_score=False)
    start = timeit.default_timer()
    clf.fit(X_train,Y_train)
    stop = timeit.default_timer()
    time.append(stop-start)
    scores.append({
      'model_name':model_name,
      'best_score':clf.best_score_,
      'best_parameters':clf.best_params_
    }
    )
  a = []
  b = []
  final_scores = list(scores)
  for i in range(len(final_scores)):
    a.append(list(final_scores[i].values()))
  for i in range(len(a)):
    b.append(a[i][1])
  maxi_ind = -1
  for i in range(len(b)):
    if(b[i] == max(b)):
      maxi_ind = i
      break
  model = algo[maxi_ind]
  model.fit(X_train,Y_train)
  X_test_prediction = model.predict(X_test)
  ml_final_result_accuracy = accuracy_score(X_test_prediction,Y_test)


  scores_t = []
  for i in range(len(scores)):
    mod = model_params[keys[i]]['model']
    print(mod)
    if scores[i]['best_parameters'] == {}:
      mod.fit(X_train, Y_train)
      scores_t.append(cross_val_score(mod, X, Y, cv=10))
    else:
      mod.set_params(**scores[i]['best_parameters'])
      scores_t.append(cross_val_score(mod, X, Y, cv=10))
  for i in range(len(scores_t)):
    if i ==maxi_ind:
      pass
    else:
      self_paired_ttest(2.262,scores_t[maxi_ind],scores_t[i],algo_names[i])
  input_data_as_np_array = np.asarray(input_data)
  input_data_reshaped = input_data_as_np_array.reshape(1,-1)
  prediction = model.predict(input_data_reshaped)
  prediction[0]
```
